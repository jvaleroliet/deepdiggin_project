```python
def find_detail(text):
    return driver.find_elements(By.XPATH, ".//dl[@class='{}']/dd/div/div/span".format(text))[0].text
```

```python
don799[["record_condition", "cover_condition", "comment", "type", "link_audio"]]=""
```

```python
driver = webdriver.Firefox()
link = don799["Link"].tolist()[95]
driver.get(link)
```

```python
# driver = webdriver.Firefox()

i=7
for link in don799["Link"].tolist()[i:]:

    # if i==0:
    #     # Wait for the GDPR banner accept button to be clickable
    #     accept_button = WebDriverWait(driver, 20).until(
    #         EC.element_to_be_clickable((By.ID, "gdpr-banner-accept"))
    #     )
    #     print("OK")
    #     # Scroll to the accept button
    #     # driver.execute_script("arguments[0].scrollIntoView();", accept_button)

    #     # Click the accept button
    #     accept_button.click()
    driver.get(link)
    sleep(1)
    try:
        don799.loc[i, "record_condition"] = find_detail('ux-labels-values ux-labels-values--inline col-6 ux-labels-values--recordGrading')
    except:
        print("no_record_condition")
    try:
        don799.loc[i, "cover_condition"] = find_detail('ux-labels-values ux-labels-values--inline col-6 ux-labels-values--sleeveGrading')
    except:
        print("nosleev")
    try:
        don799.loc[i, "comment"] = find_detail("ux-labels-values ux-labels-values--inline col-6 ux-labels-values--sellerNotes")
    except:
        print("noccoment")
    try:
        don799.loc[i, "type"] = find_detail("ux-labels-values ux-labels-values--inline col-6 ux-labels-values--type")
    except:
        print("notype")
    driver.switch_to.frame("desc_ifr")
    try:
        list = driver.find_elements(By.XPATH,"//*[contains(text(), 'http://www.ecommerceplayer.com')]")[1].get_attribute("text").split("\\u003C")
        link_audio = [x for x in list if "http://www.ecommerceplayer.com" in x][0].split('">')[-1]
        don799.loc[i, "link_audio"] = link_audio
    except:
        print("NoLink")
    print(i)
    i+=1
```
